```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

# I opened the .data file by notepad and then replace all empty space into comma then converted file in to csv data type.

```python
data =pd.read_csv('https://raw.githubusercontent.com/harvey11230/Week-8/master/auto-mpg.csv', header=None)
```

```python
data.head()
```

# assign columns to the dataframe

```python
data.columns = ['mpg','cylinders','displacement','horsepower','weight','acceleration','model year','origin','car name','none']
```

# deleted the uncessary column

```python
del data['none']
```

```python
data.head()
```

# locate records with the value of '?' in horsepower

```python
data.loc[data['horsepower']=='?']
```

# replace '?' in horsepower to 0

```python
data.loc[(data.horsepower == '?'),'horsepower'] = 0
```

# Converted origin column values are 1, 2, and 3 into 'USA', 'Asia', and 'Europe'

```python
data.loc[(data.origin == 1), 'origin'] = 'USA'
```

```python
data.loc[(data.origin == 2), 'origin'] = 'Asia'
```

```python
data.loc[(data.origin == 3), 'origin'] = 'Europe'
```

```python
data
```

# extract USA, Asia, Europe from the dataset into three data tables for increase the query speed.

```python
USA = data.loc[data['origin']=='USA']
```

```python
Asia = data.loc[data['origin']=='Asia']
```

```python
Europe = data.loc[data['origin']=='Europe']
```

# Set cylinders as index

```python
USA.set_index(['cylinders'], inplace=True)
```

```python
Asia.set_index(['cylinders'], inplace=True)
```

```python
Europe.set_index(['cylinders'], inplace=True)
```

```python
USA_cylinders = USA['origin']
```

```python
USA_cylinders = USA_cylinders.groupby(USA.index)
```

```python
Asia_cylinders = Asia['origin']
```

```python
Asia_cylinders = Asia_cylinders.groupby(Asia.index)
```

```python
Europe_cylinders = Europe['origin']
```

```python
Europe_cylinders = Europe_cylinders.groupby(Europe.index)
```

# Count vehicles in different cylinders in USA.

```python
USA_cylinders.value_counts().plot(kind='bar')
plt.title('distribution for cylinders in USA')
```

# Count vehicles in different cylinders in Asia.

```python
Asia_cylinders.value_counts().plot(kind='bar')
plt.title('distribution for cylinders in Asia')
```

# Count vehicles in different cylinders in Europe.

```python
Europe_cylinders.value_counts().plot(kind='bar')
plt.title('distribution for cylinders in Europe')
```

# USA citizen had more vehicles than Asia and Europe. And also, USA had higher number of cars in different type of cylinders than Asia and Europe either.

```python
sns.set(color_codes=True)
```

```python
sns.scatterplot(x="weight",y="horsepower", data= USA)
plt.title('USA')
```

```python
sns.scatterplot(x="weight",y="horsepower", data= Asia)
plt.title("Asia")
```

```python
sns.scatterplot(x="weight",y="horsepower", data= Europe)
plt.title("Europe")
```
